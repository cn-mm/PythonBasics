```python
#MODULES
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Image analysis and video analysis paradigms are identical
#Videos work in frames per second, so individual images are being procesed

#Using cv2
img = cv2.imread('SF.jpg',cv2.IMREAD_GRAYSCALE)
cv2.imshow('image',img)   #cv2.imshow('title',imageFile)
cv2.waitKey(0)            #press any key to close window
cv2.destroyAllWindows()
```

```python
#Saving image in directory

cv2.imwrite('gray_fox.png', img)
```

```python
#Using matplotlib 

plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.show()
```

```python
# Plotting things on top of the image 

plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.plot([200,300,400],[100,200,300],'g', linewidth=3)
plt.show()
```
