```python
#MODULES
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#IMAGE LOADING
img1 = cv2.imread('ptemplate.jpg',0) #0 stands for grayscale conversion
img2 = cv2.imread('pmain.jpg',0)
```

```python
orb = cv2.ORB_create() 

#CREATING KEY POINTS AND THEIR DESCRIPTORS 
kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)

#BF MATCHER OBJECT
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True) #Norm_Hamming is a Method

#MATCHES OF DECRIPTIONS AND SORTING THEM 
matches = bf.match(des1,des2)
matches = sorted(matches, key = lambda x:x.distance)

#DISPLAY
img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[:10],None, flags=2)
plt.imshow(img3)
plt.show()
```
